```python
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Create a directory to store the PDF files if it doesn't exist
folder_name = "apple_filings1"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

j = -8
for i in range(2003, 2023):
    file_name = os.path.join(folder_name, f"apple_{i}.pdf")
    latest_filing = filings[j].text()  # Assuming filings is a list or dictionary
    j -= 1

    # Create a PDF document
    c = canvas.Canvas(file_name, pagesize=letter)
    c.setFont("Helvetica", 12)

    # Write the text to the PDF using utf-8 encoding
    c.drawString(100, 700, latest_filing.encode('latin-1', 'replace').decode('latin-1'))

    # Save the PDF
    c.save()

print("PDF files created successfully in folder 'apple_filings'")
```
